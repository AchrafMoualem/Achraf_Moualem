# ðŸŒŸ Exercise 1 : Matrix Operations

```python
import numpy as np

matrix = np.array([[2, 3, 8], [9, 4, 1], [7, 7, 6]])
matrix 
```

```python
determinant = np.linalg.det(matrix)
inverse = np.linalg.inv(matrix)
print("the determinant : ", determinant)
print("the inverse : ", inverse)
```

```python
print(matrix.dot(inverse))
```

# ðŸŒŸ Exercise 4 : Data Manipulation with NumPy and Pandas


```python
import pandas as pd

df = pd.DataFrame(np.random.rand(8, 4), columns= [1,2,3,4])
df
```

```python
df_filtered_pd = df[(df[1]>0.5) & (df[2]<0.5)].mean()
df_filtered_pd
```

```python
df_np = df.to_numpy()
df_filtered_np = np.mean(df_np[(df_np[:, 0] > 0.5) & (df_np[:, 1] < 0.5)], axis=0)
df_filtered_np
```

```python
df_filtered_sum_pd = df[(df[1]>0.5) & (df[2]<0.5)].sum()
df_filtered_sum_pd
```

```python
df_filtered_sum_np = np.sum(df_np[(df_np[:, 0] > 0.5) & (df_np[:, 1] < 0.5)], axis=0)
df_filtered_sum_np
```

# ðŸŒŸ Exercise 6 : Basic Hypothesis Testing


```python
from scipy import stats
# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

t_statistic, p_value = stats.ttest_rel(productivity_after, productivity_before)

alpha = 0.05
if p_value < alpha:
    print(f"Reject H0: Training had significant effect (p={p_value:.4f})")
else:
    print(f"Fail to reject H0: No significant effect (p={p_value:.4f})")
```

# ðŸŒŸ Exercise 8 : Time Series Data Manipulation


```python
import numpy as np
import pandas as pd
# Generate time series for 2023 (daily data)
dates = np.arange('2023-01', '2023-12', dtype='datetime64[D]')
values = np.random.randn(len(dates))
ts = pd.Series(values, index=dates)

# Slice for different quarters
jan_mar = ts['2023-01':'2023-03']
apr_jun = ts['2023-04':'2023-06']
jul_sep = ts['2023-07':'2023-09']
oct_dec = ts['2023-10':'2023-12']

print("Jan-Mar shape:", jan_mar.shape)
print("Apr-Jun shape:", apr_jun.shape)
print("Jul-Sep shape:", jul_sep.shape)
print("Oct-Dec shape:", oct_dec.shape)
```
