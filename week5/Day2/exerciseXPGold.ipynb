# Exercise 1 : Advanced Matrix Operations


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
matrix_5x5 = np.random.randint(1, 101, size=(5, 5))
matrix_5x5
```

```python
eigenvalues, eigenvectors = np.linalg.eig(matrix_5x5)
print("eigenvalues : ", eigenvalues)
print("eigenvectors : ", eigenvectors)
```

```python
normalized_matrix = (matrix_5x5 - np.min(matrix_5x5)) / (np.max(matrix_5x5) - np.min(matrix_5x5))
print("Normalized Matrix : \n", normalized_matrix)
```

# Exercise 2 : Statistical Distribution Analysis


```python
array_normal = np.random.normal(loc=0.0, scale=1, size=(100,))
plt.hist(array_normal, bins=15, edgecolor='black')
plt.show()
```

```python
from scipy.stats import skew, kurtosis
print("Skewness : ", skew(array_normal))
print("Kurtosis : ", kurtosis(array_normal))    
```

# Exercise 3 : Time Series Forecasting


```python
from sklearn.linear_model import LinearRegression
monthly_sales_data = np.random.randint(100, 500, size=12)
X = np.array(range(1, 13)).reshape(-1, 1)
y = monthly_sales_data
model = LinearRegression()
model.fit(X, y)
predicted_sales = model.predict(X)
print("Predicted Sales : ", predicted_sales)
```

# Exercise 4 : Pandas Data Aggregation


```python
# Creating a sample Pandas DataFrame
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product A', 'Product B'],
    'Sales': [200, 150, 300, 250, 180],
    'Month': ['January', 'February', 'March', 'April', 'May']
}
df = pd.DataFrame(data)
grouped_sales = df.groupby('Product')['Sales'].sum()
print("Total Sales by Product : \n", grouped_sales)
```

# Exercise 5 : Complex Data Visualization


```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y_line = np.sin(x)
y_scatter = np.random.randn(50)
categories = ['A', 'B', 'C', 'D']
values = np.random.randint(10, 50, 4)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Line plot
axes[0,0].plot(x, y_line, 'r-', linewidth=2)
axes[0,0].set_title('Line Plot')
axes[0,0].grid(True)

# Scatter plot
axes[0,1].scatter(range(50), y_scatter, alpha=0.5, c='green')
axes[0,1].set_title('Scatter Plot')

# Bar plot
axes[1,0].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axes[1,0].set_title('Bar Plot')

# Histogram
axes[1,1].hist(y_scatter, bins=15, edgecolor='black', alpha=0.7)
axes[1,1].set_title('Histogram')

plt.suptitle('Multi-Plot Layout with NumPy Data', fontsize=16, y=1.02)
plt.show()
```
