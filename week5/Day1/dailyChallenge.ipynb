# Daily Challenge: NumPy, Pandas, and Matplotlib Integration

### 1. Data Preparation:

```python
import numpy as np

temperature = np.random.uniform(-5, 35, 120)
temperature_reshaped = temperature.reshape(10, 12)
cities = ['Casablanca', 'Rabat', 'Mohammedia', 'Kenitre', 'El jadida', 'Settat', 'Azemour', 'Marrakesh', 'Ifrane', 'Fes']
months = ['January', 'February', 'Mars', 'April', 'May', 'Juni', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
import pandas as pd
meteo = pd.DataFrame(temperature_reshaped , index= cities, columns= months)
meteo.head()
```

### 2. Data Analysis:

```python
overall_annual_average = meteo.values.mean()
print(f"Overall annual average temperature: {overall_annual_average:.2f}°C")
```

```python
annual_average = meteo.mean(axis=1)
annual_average
```

```python
highest_avg = annual_average.idxmax()
lowest_avg = annual_average.idxmin()
print("the city with the highest average temperature is ", highest_avg)
print("the city with the lowest average temperature is ", lowest_avg)
```

### 3. Data Visualization:



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Simple line plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=meteo.T)  # Transpose to plot cities as lines

plt.title('Monthly Temperature by City')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.xticks(range(12), months, rotation=45)
plt.legend(cities, bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

as we can see in this plot, the highest temperature is marked in January and November by Marrakesh and Rabat resp. and the lowest is marked in Novemberand, Juni and September by El jadida, Marrakesh and Rabat resp. and about Casablanca we can see from the plot that is regular, its temperature is a little high in between January and July and start to reduces from July until November when it start to get higher again.
