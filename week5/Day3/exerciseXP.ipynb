# ðŸŒŸ Exercise 1: Basic Usage of SciPy

```python
import scipy.stats as stats
print(f'the scipy version is {scipy.__version__}')
```

# ðŸŒŸ Exercise 2: Descriptive Statistics

```python
import numpy as np

data = [12, 15, 13, 12, 18, 20, 22, 21]

mean = stats.tmean(data)
mediam = np.median(data)
var = stats.tvar(data)
std_dev = stats.tstd(data)

print(f'the mean is {mean}')
print(f'the median is {mediam}')
print(f'the variance is {var}')
print(f'the standard deviation is {std_dev}')
```

# ðŸŒŸ Exercise 3: Understanding Distributions

```python
import matplotlib.pyplot as plt

mean, std = 50, 10
gaussian_dist = stats.norm(mean, std)
print(gaussian_dist.pdf(50))

# Plotting Gaussian Distribution
x_gaussian = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf_gaussian = gaussian_dist.pdf(x_gaussian)
plt.figure(figsize=(10, 6))
plt.plot(x_gaussian, pdf_gaussian)
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()
```

# ðŸŒŸ Exercise 4: T-Test Application

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = stats.ttest_ind(data1, data2)
print(f't-statistic: {t_stat}')
print(f'p-value: {p_value}')
```

# ðŸŒŸ Exercise 5: Linear Regression Analysis

```python
house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]  # No commas in the integers.

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)
print(f'slope: {slope}')
print(f'intercept: {intercept}')
```

# ðŸŒŸ Exercise 6: Understanding ANOVA

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

anova_result = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print(f'F-statistic: {anova_result.statistic}')
print(f'p-value: {anova_result.pvalue}')
```

if the p_value were greater than 0.05 that will mean that there is a significant difference between the 3 datasets means
