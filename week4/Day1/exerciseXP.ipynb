# ðŸŒŸ Exercise 1 : Understanding Data Visualization


Data visualization transforms numbers into visual formats (charts, graphs) making patterns and trends instantly recognizable - our brains process images much faster than raw data tables.

A line graph specifically shows how values change over time by connecting data points with lines, making it easy to spot trends, compare multiple datasets, and identify patterns at a glance.

# ðŸŒŸ Exercise 2 : Creating a Line Plot for Temperature Variation


```python
import matplotlib.pyplot as plt
import pandas as pd

df = [72, 74, 76, 80, 82, 78, 75]
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

plt.plot(days, df, marker='o')
plt.title('Weekly Temperature')
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (Â°F)')
plt.show()
```

# ðŸŒŸ Exercise 3 : Visualizing Monthly Sales with a Bar Chart


```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales = [1500, 1800, 2200, 2400, 2000, 2300, 2500, 2700, 3000, 3200, 3100, 4000]

plt.bar(months, monthly_sales, color='skyblue')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.show()
```

# ðŸŒŸ Exercise 4 : Data Visualisation

```python
sales = pd.read_csv('sales_data.csv')
sales.head()
```

```python
qtq_product = sales['quantity'].sum()
qtq_product
```

```python
highest = sales['category'][sales['revenue'].idxmax()]
highest
```

```python
avg_revenue = sales['revenue'].mean()
avg_revenue
```

```python
sales['date'] = pd.to_datetime(sales['date'])
sales['quarter'] = sales['date'].dt.quarter
plt.bar(sales['quarter'], sales['revenue'], color='orange')
plt.title('Quarterly Revenue')
plt.xlabel('Quarter')
plt.ylabel('Revenue ($)')
plt.xticks([1, 2, 3, 4], ['Q1', 'Q2', 'Q3', 'Q4'])

plt.show()
```

# ðŸŒŸ Exercise 5 : Data Visualisation Using MatPlotLib


```python
x = range(-10, 11)
y = [i**2 for i in x]
plt.plot(x, y, color='blue')
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]
plt.bar(products, sales, color='green')
plt.title('Product Sales')  
plt.xlabel('Products')
plt.ylabel('Number of Sales')
plt.show()
```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
favorites = [40, 30, 20, 10]
plt.pie(favorites, labels=fruits, autopct='%1.1f%%', colors=['red', 'yellow', 'pink', 'brown'])
plt.title('Favorite Fruits Distribution')
plt.legend(fruits, title="Fruits", loc="best")
plt.show()               
```
