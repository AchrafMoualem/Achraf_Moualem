## Daily Challenge: Hands-On Image Transformation and Visualization



```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = 'Flower Color Images Dataset/flowers/flowers/19_010.png'
original_image = Image.open(image_path)
```

```python
plt.imshow(original_image)
```

```python
from scipy.ndimage import rotate
def rotate_image_30_degrees(image):
    return image.rotate(30, expand=True, resample=Image.BICUBIC)
```

```python
rotate_image_30_degrees(original_image)
```

```python
from PIL import ImageOps
horizontal_flip = ImageOps.mirror(original_image)
vertical_flip = ImageOps.flip(original_image)
```

```python
horizontal_flip.show()
```

```python
vertical_flip.show()
```

```python
width, height = original_image.size
new_width = int(width * 1.2)
new_height = int(height * 1.2)

# Resize with high-quality resampling
zoomed_image = original_image.resize((new_width, new_height), resample=Image.LANCZOS)

zoomed_image.show()
```
