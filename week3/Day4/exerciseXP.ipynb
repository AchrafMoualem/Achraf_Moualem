# ðŸŒŸ Exercise 2: Creating a Line Plot for Temperature Variation

```python
import matplotlib.pyplot as plt

days = ['Monday', 'Tuesady', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
temperature = [72, 74, 76, 80, 82, 78, 75]

plt.plot(days, temperature)
plt.title("Temperature of each day in the week")
plt.xlabel("day")
plt.ylabel("Temperature (Â°F)")
plt.show()
```

# ðŸŒŸ Exercise 3: Visualizing Monthly Sales with a Bar Chart

```python
months = ['January', 'February', 'March', 'April', 'May']
sales_values = [5000, 5500, 6200, 7000, 7500]

plt.bar(months, sales_values)
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.show()
```

# ðŸŒŸ Exercise 4: Visualizing the Distribution of CGPA

```python
import pandas as pd

df = pd.read_csv('Student Mental health.csv')
df.head()
```

```python
import seaborn as sns

sns.histplot(df['What is your CGPA?'], bins=15 , color='green')
plt.title('the CGPA categories')
plt.show()
```

# ðŸŒŸ Exercise 5: Comparing Anxiety Levels Across Different Genders


```python
sns.barplot(x = df['Choose your gender'], y = df['Do you have Anxiety?'], data = df, palette='Blues')
plt.title("the proportion of students experiencing anxiety across different genders")
plt.show()
```

# ðŸŒŸ Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
df['Do you have Panic attack?'] = df['Do you have Panic attack?'].replace({'Yes':1 , 'No':0})
```

```python
df.head()
```

```python
sns.scatterplot(x='Age', y='Do you have Panic attack?', data=df, hue='Do you have Panic attack?')
plt.title('the relationship between students age and the occurrence of panic attacks.')
plt.xlabel("Age")
plt.ylabel("panic attack")
plt.show()
```
